```python
ETHTRY_sell= data_frame[(data_frame["Market"]=="ETHTRY") & (data_frame["OrderType"]=="sell")]["Total"].sum()
ETHTRY_buy= data_frame[(data_frame["Market"]=="ETHTRY") & (data_frame["OrderType"]=="buy")]["Total"].sum()
USDTTRY_sell= data_frame[(data_frame["Market"] =="USDTTRY") & (data_frame["OrderType"] =="sell")]["Total"].sum()
USDTTRY_buy= data_frame[(data_frame["Market"] =="USDTTRY") & (data_frame["OrderType"] =="buy")]["Total"].sum()
BTCTRY_sell= data_frame[(data_frame["Market"] =="BTCTRY") & (data_frame["OrderType"]=="sell")]["Total"].sum()
BTCTRY_buy= data_frame[(data_frame["Market"]=="BTCTRY") &(data_frame["OrderType"]=="buy")]["Total"].sum()
ETHUSDT_sell= data_frame[(data_frame["Market"] =="ETHUSDT") & (data_frame["OrderType"] =="sell")]["Total"].sum()
ETHUSDT_buy = data_frame[(data_frame["Market"] == "ETHUSDT") & (data_frame["OrderType"] =="buy")]["Total"].sum()
XLMTRY_sell= data_frame[(data_frame["Market"] == "XLMTRY") & (data_frame["OrderType"] == "sell")]["Total"].sum()
XLMTRY_buy = data_frame[(data_frame["Market"] == "XLMTRY") & (data_frame["OrderType"]== "buy") ]["Total"].sum()
DOTTRY_sell= data_frame[(data_frame["Market"] == "DOTTRY") & (data_frame["OrderType"]== "sell")]["Total"].sum()
DOTTRY_buy = data_frame[(data_frame["Market"] == "DOTTRY") & (data_frame["OrderType"] == "buy") ]["Total"].sum()
BTCUSDT_sell = data_frame[(data_frame["Market"]== "BTCUSDT") & (data_frame["OrderType"] == "sell")]["Total"].sum()
BTCUSDT_buy = data_frame[(data_frame["Market"] == "BTCUSDT") & (data_frame["OrderType"] == "buy")]["Total"].sum()
ADATRY_sell= data_frame[(data_frame["Market"]=="ADATRY") & (data_frame["OrderType"]== "sell")]["Total"].sum()
ADATRY_buy = data_frame[(data_frame["Market"] =="ADATRY") & (data_frame["OrderType"] == "buy")]["Total"].sum()
```

```python
ETHTRY_total_revenue= -(ETHTRY_buy+ETHTRY_sell)
USDTTRY_total_revenue= -(USDTTRY_buy+USDTTRY_sell)
BTCTRY_total_revenue= -(BTCTRY_buy+BTCTRY_sell)
ETHUSDT_total_revenue= -(ETHUSDT_buy+ETHUSDT_sell)
XLMTRY_total_revenue= -(XLMTRY_buy+XLMTRY_sell)
DOTTRY_total_revenue= -(DOTTRY_buy+DOTTRY_sell)
BTCUSDT_total_revenue= -(BTCUSDT_buy+BTCUSDT_sell)
ADATRY_total_revenue= -(ADATRY_buy+ADATRY_sell)
```

```python
x= np.array([ETHTRY_total_revenue, USDTTRY_total_revenue,BTCTRY_total_revenue,ETHUSDT_total_revenue,XLMTRY_total_revenue,DOTTRY_total_revenue,BTCUSDT_total_revenue,ADATRY_total_revenue])
```
